#### Membuang kolom tidak penting untuk tujuan Modeling

```python
# remove kolom tidak penting
deleted_columns = [
    'Tanggal_Post', 
    'Berlaku_Hingga', 
    'Perusahaan', 
    'Website_Perusahaan', 
    'Logo_Perusahaan', 
    'Alamat', 
    'Latitude',
    'Longitude',
    'Gaji_Currency',
    'Gaji_Unit',
    'Tentang_Perusahaan',
    'Benefit',
    'Apply_Langsung',
    'Link'
]

df_1 = clean_job.copy()
df_2 = clean_job.copy()

# hapus kolom yang tidak penting
df_1.drop(columns=deleted_columns, inplace=True)
df_1.columns.to_list()
```

```python
# cek apakah ada duplicate
df_1.duplicated().sum()
```

```python
df_2.duplicated().sum()
```

```python
mask_duplicate = df_1.duplicated()

# hapus duplicate
df_1 = df_1[~mask_duplicate]

df_2 = df_2[~mask_duplicate]

df_2.columns.to_list()
```

```python
df_1.info()
```

```python
df_2.info()
```

```python
print(df_1.isnull().sum())
print(df_2.isnull().sum())

print(df_1.duplicated().sum())
print(df_2.duplicated().sum())
```

```python
# masih membutuhkan index sepertinya, jadi tidak perlu drop index
df_1.to_csv('data_untuk_modeling.csv', index=True)
df_2.to_csv('data_untuk_raw_backend.csv', index=True)
```
